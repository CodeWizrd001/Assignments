```python
# Hill Cipher 

def hillEnc(string,*args,**kwargs) :
    key = kwargs.get("key","KEYZ")
    m = int(len(alphaDictCaps) // 2)
    
    # Caps Implementation
    key = key.upper()
    string = string.upper()
    string = string.replace(" ","")
    
    y = len(string) 
    x = len(key)
    n = int(math.sqrt(x))
    if n**2 != x :
        raise KeyException("Key not square")
    
    # Make KeyMatrix
    keyList = [alphaDictCaps[i] for i in key] 
    keyMat = np.array(keyList)
    keyMat.resize((n,n))
    # print("Key") 
    # print(keyMat)
    
    # Pad string if needed 
    nRows = math.ceil(y/n)
    while len(string) < n * nRows :
        string += "Z"
    
    # Make String Matrix
    strList = [alphaDictCaps[i] for i in string]
    strMat = np.array(strList)
    strMat.resize((nRows,n))
    # print("String")
    # print(strMat)
    
    cipherMat = np.matmul(strMat,keyMat) % m
    # print("Cipher")
    # print(cipherMat)
    cipherList = cipherMat.flatten()
    
    cipherText = ''.join([alphaDictCaps[i] for i in cipherList])
    return cipherText

def hillDec(string,*args,**kwargs) :
    key = kwargs.get("key","KEYZ")
    m = int(len(alphaDictCaps) // 2)
    
    # Caps Implementation
    key = key.upper()
    string = string.upper()
    string = string.replace(" ","")
    
    y = len(string) 
    x = len(key)
    n = int(math.sqrt(x))
    if n**2 != x :
        raise KeyException("Key not square")
    
    # Make KeyMatrix
    keyList = [alphaDictCaps[i] for i in key] 
    keyMat = np.array(keyList)
    keyMat.resize((n,n))
    keyMatInv = np.array(Matrix(keyMat).inv_mod(m))
    # print("Key") 
    # print(keyMat)
    # print("Key Inverse") 
    # print(keyMatInv)
    
    # Make String Matrix
    nRows = math.ceil(y/n)
    strList = [alphaDictCaps[i] for i in string]
    strMat = np.array(strList)
    strMat.resize((nRows,n))
    # print("String")
    # print(strMat)
    
    plainMat = np.matmul(strMat,keyMatInv).astype(int) % m
    # print("Plain")
    # print(plainMat)
    plainList = plainMat.flatten()
    
    plainText = ''.join([alphaDictCaps[i] for i in plainList])
    return plainText
```

```python
# Transposition Cipher

def transEncKeylessRF(string) :
    m = math.ceil(len(string) / 2)
    while len(string) < 2*m :
        string += " "
    a = np.array(list(string))
    a = a.reshape((2,m))
    a = a.T
    a = a.flatten()
    output = ''.join(a)
    return output

def transDecKeylessRF(string) :
    m = math.ceil(len(string) / 2)
    while len(string) < 2*m :
        string += " "
    a = np.array(list(string))
    a = a.reshape((m,2))
    a = a.T
    a = a.flatten()
    output = ''.join(a)
    return output

def transCipherKeyless(string) :
    m = math.ceil(math.sqrt(len(string)))
    while len(string) < m*m :
        string += "."
    a = np.array(list(string))
    a = a.reshape((m,m))
    a = a.T
    a = a.flatten()
    output = ''.join(a)
    return output

def transKeyEnc(string,key) :
    m = len(key)
    while len(string) % m != 0 :
        string += " "
    chunks = [string[i*m:(i+1)*m] for i in range(int(len(string)//m))]
    ans = [[' ']*m for i in chunks]
    for i in range(len(ans)) :
        for j in range(m) :
            ans[i][j] = chunks[i][key[j]]
        ans[i] = ''.join(ans[i])
    cipherText = ''.join(ans)
    return cipherText

def transKeyDec(string,key) :
    m = len(key)
    while len(string) % m != 0 :
        string += " "
    chunks = [string[i*m:(i+1)*m] for i in range(int(len(string)//m))]
    ans = [[' ']*m for i in chunks]
    for i in range(len(ans)) :
        for j in range(m) :
            ans[i][key[j]] = chunks[i][j]
        ans[i] = ''.join(ans[i])
    plainText = ''.join(ans)
    return plainText
```

```python
# Viganere Cipher

vigTable = {}
for i in range(len(upper)) :
    vigTable[upper[i]] = {}
    for j in range(len(upper)) :
        vigTable[upper[i]][upper[j]] = upper[(i+j)%26]

def vigInv(ciphChar,keyChar) :
    for i in vigTable :
        if vigTable[i][keyChar] == ciphChar :
            return i
    return None

def vigEnc(string,key="KEY",*args,**kwargs) :
    # Caps Implementation
    key = key.upper()
    string = string.upper()
    string = string.replace(" ","")
    
    t = key
    while len(key) < len(string) :
        key += t
    key = key[:len(string)]
    cipherText = ""
    for i in range(len(string)) :
        cipherText += vigTable[string[i]][key[i]]
    return cipherText

def vigDec(string,key="KEY",*args,**kwargs) :
    # Caps Implementation
    key = key.upper()
    string = string.upper()
    string = string.replace(" ","")
    
    t = key
    while len(key) < len(string) :
        key += t
    key = key[:len(string)]
    plainText = ""
    for i in range(len(string)) :
        plainText += vigInv(string[i],key[i])
    return plainText
```
