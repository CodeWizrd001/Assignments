```python
def power(x, y, m) : 
    if (y == 0) : 
        return 1
    p = power(x, y // 2, m) % m 
    p = (p * p) % m 
    print(p)
    if(y % 2 == 0) : 
        return p  
    else :  
        return ((x * p) % m)

def modInverse(a, m) : 
    g = gcd(a, m) 
    if (g != 1) : 
        return -1 
    else :
        return power(a, m - 2, m)
    
def bruteInverse(a,m) :
    if gcd(a,m) != 1 :
        return -1 
    for i in range(26) :
        if (i*a)%m == 1 :
            return i

def getRandomValue(*args,**kwargs) :
    return random.randint(0,100)

def getRandomKey() :
    temp = list(upper)
    temp.sort(key=getRandomValue)
    return ''.join(temp)
```

```python
# Affine Cipher

def affineEnc(string,a=1,b=0,*args,**kwargs) :
    # All Caps or not
    caps = kwargs.get('caps',True)
    # Dictionary of characters
    if caps :
        string = string.upper()
        aD = kwargs.get('aD',alphaDictCaps)
    else :
        aD = kwargs.get('aD',alphaDict)
    m = len(aD) / 2
    m = int(m)
    if gcd(m,a) != 1 :
        raise ParameterError("a and m Not CoPrime")
    cipherText = "".join([aD[(a*aD[i] + b ) % m] for i in string])
    return cipherText

def affineDec(string,a=1,b=0,*args,**kwargs) :
    # All Caps or not
    caps = kwargs.get('caps',True)
    # Dictionary of characters
    if caps :
        string = string.upper()
        aD = kwargs.get('aD',alphaDictCaps)
    else :
        aD = kwargs.get('aD',alphaDict)
    m = len(aD) / 2
    m = int(m)
    if gcd(m,a) != 1 :
        raise ParameterError("a and m Not CoPrime")
    a = bruteInverse(a,m)
    plainText = "".join([aD[(a*(aD[i] - b )) % m] for i in string])
    return plainText
```

```python
# Hill Cipher 

def hillEnc(string,*args,**kwargs) :
    key = kwargs.get("key","KEYZ")
    m = int(len(alphaDictCaps) // 2)
    
    # Caps Implementation
    key = key.upper()
    string = string.upper()
    string = string.replace(" ","")
    
    y = len(string) 
    x = len(key)
    n = int(math.sqrt(x))
    if n**2 != x :
        raise KeyException("Key not square")
    
    # Make KeyMatrix
    keyList = [alphaDictCaps[i] for i in key] 
    keyMat = np.array(keyList)
    keyMat.resize((n,n))
    
    # Pad string if needed 
    nRows = math.ceil(y/n)
    while len(string) < n * nRows :
        string += "Z"
    
    # Make String Matrix
    strList = [alphaDictCaps[i] for i in string]
    strMat = np.array(strList)
    strMat.resize((nRows,n))
    
    cipherMat = np.matmul(strMat,keyMat) % m
    cipherList = cipherMat.flatten()
    
    cipherText = ''.join([alphaDictCaps[i] for i in cipherList])
    return cipherText

def hillDec(string,*args,**kwargs) :
    key = kwargs.get("key","KEYZ")
    m = int(len(alphaDictCaps) // 2)
    
    # Caps Implementation
    key = key.upper()
    string = string.upper()
    string = string.replace(" ","")
    
    y = len(string) 
    x = len(key)
    n = int(math.sqrt(x))
    if n**2 != x :
        raise KeyException("Key not square")
    
    # Make KeyMatrix
    keyList = [alphaDictCaps[i] for i in key] 
    keyMat = np.array(keyList)
    keyMat.resize((n,n))
    keyMatInv = np.array(Matrix(keyMat).inv_mod(m))
    
    # Make String Matrix
    nRows = math.ceil(y/n)
    strList = [alphaDictCaps[i] for i in string]
    strMat = np.array(strList)
    strMat.resize((nRows,n))
    
    plainMat = np.matmul(strMat,keyMatInv).astype(int) % m
    plainList = plainMat.flatten()
    
    plainText = ''.join([alphaDictCaps[i] for i in plainList])
    return plainText
```

```python
# Substitution Cipher

def subEnc(string,key,*args,**kwargs) :
    string = string.upper()
    string = string.replace(" ","")
    keyMap = {}
    for i in range(len(key)) :
        keyMap[upper[i]] = key[i]
    cipherText = ''.join([keyMap[i] for i in string])
    return cipherText

def subDec(string,key,*args,**kwargs) :
    string = string.upper()
    string = string.replace(" ","")
    keyMap = {}
    for i in range(len(key)) :
        keyMap[key[i]] = upper[i]
    plainText = ''.join([keyMap[i] for i in string])
    return plainText
```

```python
# Transposition Cipher

def transEncKeylessRF(string) :
    m = math.ceil(len(string) / 2)
    while len(string) < 2*m :
        string += " "
    a = np.array(list(string))
    a = a.reshape((2,m))
    a = a.T
    a = a.flatten()
    output = ''.join(a)
    return output

def transDecKeylessRF(string) :
    m = math.ceil(len(string) / 2)
    while len(string) < 2*m :
        string += " "
    a = np.array(list(string))
    a = a.reshape((m,2))
    a = a.T
    a = a.flatten()
    output = ''.join(a)
    return output

def transCipherKeyless(string) :
    m = math.ceil(math.sqrt(len(string)))
    while len(string) < m*m :
        string += "."
    a = np.array(list(string))
    a = a.reshape((m,m))
    a = a.T
    a = a.flatten()
    output = ''.join(a)
    return output

def transKeyEnc(string,key) :
    m = len(key)
    while len(string) % m != 0 :
        string += " "
    chunks = [string[i*m:(i+1)*m] for i in range(int(len(string)//m))]
    ans = [[' ']*m for i in chunks]
    for i in range(len(ans)) :
        for j in range(m) :
            ans[i][j] = chunks[i][key[j]]
        ans[i] = ''.join(ans[i])
    cipherText = ''.join(ans)
    return cipherText

def transKeyDec(string,key) :
    m = len(key)
    while len(string) % m != 0 :
        string += " "
    chunks = [string[i*m:(i+1)*m] for i in range(int(len(string)//m))]
    ans = [[' ']*m for i in chunks]
    for i in range(len(ans)) :
        for j in range(m) :
            ans[i][key[j]] = chunks[i][j]
        ans[i] = ''.join(ans[i])
    plainText = ''.join(ans)
    return plainText
```

```python
# Viganere Cipher

vigTable = {}
for i in range(len(upper)) :
    vigTable[upper[i]] = {}
    for j in range(len(upper)) :
        vigTable[upper[i]][upper[j]] = upper[(i+j)%26]

def vigInv(ciphChar,keyChar) :
    for i in vigTable :
        if vigTable[i][keyChar] == ciphChar :
            return i
    return None

def vigEnc(string,key="KEY",*args,**kwargs) :
    # Caps Implementation
    key = key.upper()
    string = string.upper()
    string = string.replace(" ","")
    
    t = key
    while len(key) < len(string) :
        key += t
    key = key[:len(string)]
    cipherText = ""
    for i in range(len(string)) :
        cipherText += vigTable[string[i]][key[i]]
    return cipherText

def vigDec(string,key="KEY",*args,**kwargs) :
    # Caps Implementation
    key = key.upper()
    string = string.upper()
    string = string.replace(" ","")
    
    t = key
    while len(key) < len(string) :
        key += t
    key = key[:len(string)]
    plainText = ""
    for i in range(len(string)) :
        plainText += vigInv(string[i],key[i])
    return plainText
```

```python
# Cryptanalysis of Hill Cipher

def knownHill(cipherText,plainText) :
    m = 26
    
    # Caps Implementation
    plainText = plainText.upper()
    plainText = plainText.replace(" ","")
    cipherText = cipherText.upper()
    cipherText = cipherText.replace(" ","")
    
    assert len(plainText) == len(cipherText) , "Invalid Pair"
    
    x = len(cipherText)
    n = int(math.sqrt(x))
    assert n**2 == x , "Invalid plainText"
    
    # Make Cipher String Matrix
    cipList = [alphaDictCaps[i] for i in cipherText]
    cipMat = np.array(cipList)
    cipMat.resize((n,n))
    
    # Make Plain String Matrix
    strList = [alphaDictCaps[i] for i in plainText]
    strMat = np.array(strList)
    strMat.resize((n,n))
    
    try :
        invStrMat = Matrix(strMat).inv_mod(26)
    except :
        raise ParameterError("String Not Invertible")
    
    keyMat = np.matmul(cipMat,invStrMat).astype(int) % m
    keyList = keyMat.flatten()
    
    keyText = ''.join([alphaDictCaps[i] for i in keyList])
    return keyText
```

```python
# Cryptanalysis of Vigenere Cipher

vigTable = {}
for i in range(len(upper)) :
    vigTable[upper[i]] = {}
    for j in range(len(upper)) :
        vigTable[upper[i]][upper[j]] = upper[(i+j)%26]

def vigKey(plainChar,ciphChar) :
    for i in vigTable[plainChar] :
        if vigTable[plainChar][i] == ciphChar :
            return i

def knownVig(cipherText,plainText) :
    # Caps Implementation
    plainText = plainText.upper()
    plainText = plainText.replace(" ","")
    cipherText = cipherText.upper()
    cipherText = cipherText.replace(" ","")
    
    assert len(plainText) == len(cipherText) , "Invalid Pair"
    
    key = ""
    for i in range(len(plainText)) :
        key += vigKey(plainText[i],cipherText[i])
    return key

def listGCD(l) :
    if len(l) == 0 :
        raise ParameterException("Empty List")
    elif len(l) == 1 :
        return l[0]
    gcd = math.gcd(l[0],l[1])
    for i in l :
        gcd = math.gcd(gcd,i)
    return gcd

def statisticalVig(cipherChunk) :
    freqList = [(cipherChunk.count(i) / len(cipherChunk)) for i in cipherChunk]
    tFreqDict = {}
    for i in cipherChunk :
        try :
            tFreqDict[i] += 1 
        except :
            tFreqDict[i] = (cipherChunk.count(i) / len(cipherChunk))
    freqOrig = freqDict.values()
    temp = []
    for i in freqOrig :
        temp += i
    freqOrig = temp
    
    chunkList = list(set(cipherChunk))
    chunkListSorted = sorted(chunkList,reverse=True,key = lambda x : tFreqDict[x])
    
    chunkMap = {}
    
    for i in range(len(chunkListSorted)) :
        chunkMap[chunkListSorted[i]] = freqOrig[i]
    
    chunkToReturn = ""
    for i in cipherChunk :
        chunkToReturn += chunkMap[i]
    
    return chunkToReturn

def kasiskiTest(cipherText,top=-1) :
    repLength = 3
    repList = set()
    repDict = {}
    
    cipherText = cipherText.upper()
    
    for i in range(len(cipherText)-repLength) :
        sequence = cipherText[i:i+repLength]
        repList.add((i,sequence))
        try :
            repDict[sequence].add(i)
        except :
            repDict[sequence] = {i}
        for j in range(i,len(cipherText)-repLength) :
            if sequence == cipherText[j:j+repLength] :
                repList.add((j,sequence))
                repDict[sequence].add(i)
    
    newRepDict = {}
    for i in repDict :
        if len(repDict[i]) > 1 :
            newRepDict[i] = list(repDict[i])
    repDict = newRepDict
    
    diffs = set()
    for i in repDict :
        diffs.add(abs(repDict[i][1]-repDict[i][0]))
        if len(repDict[i]) > 2 :
            diffs.add(abs(repDict[i][2]-repDict[i][1]))
    diffs = list(diffs)
    keyBase = listGCD(diffs)
    
    i = 1
    keyLength = keyBase * i
    
    freqDict_ = freqDict
    
    possibles = []
    
    while keyLength <= len(cipherText) :
        chunks = [cipherText[i:i+keyLength] for i in range(0,len(cipherText),keyLength)]
        plainChunks = [statisticalVig(i) for i in chunks]
        possibles.append(''.join(plainChunks))
        i += 1
        keyLength = keyBase * i
    
    matchDict = {}
    matchList = []
    
    for i in possibles :
        for j in dictionary :
            if j.upper() in i :
                try :
                    matchDict[i] += len(j)
                except :
                    matchDict[i] = len(j)
                    matchList.append(i)
                
    matchList.sort(reverse=True,key = lambda x : matchDict[x])
    if top == -1 :
        return matchList
    else :
        return matchList[:top]
```

```python
# Cryptanalysis of Transposition Cipher

def factors(n) :
    for i in range(2,n) :
        if not n % i :
            yield i

def bruteTrans(cipherText,top = -1) :
    l = len(cipherText)
    
    cipherText = cipherText.upper()
    
    blockLength = 0
    
    bruteList = []
    
    for bl in factors(l) :
        if bl > 10 :
            break
        for key in permutations([i for i in range(bl)]) :
            plainText = transKeyDec(cipherText,key)
            plainText = plainText.replace(" ","")
            bruteList.append(plainText)
    
    matchDict = {}
    matchList = []
    
    for i in bruteList :
        for j in dictionary :
            if j.upper() in i :
                try :
                    matchDict[i] += len(j)
                except :
                    matchDict[i] = len(j)
    
    for i in matchDict :
        matchList.append((matchDict[i],i))
                
    matchList.sort(reverse=True,key = lambda x : matchDict[x[1]])
    if top == -1 :
        return matchList
    else :
        return matchList[:top]
```
